**Part 3: Implementation of Adaline**

Specifications: 

1. Implement the Adaline algorithm to your dataset with learning rate of 0.01 and set the tolerance to 0.015. 

2. Plot the mean squared deviation vs. epoch. Does the algorithm converge fast? How many epochs was executed?

3. Classify the data set into two groups: Setosa and Versicolor, where now the output is the predicted output in the Adaline algorithm. 
Note: Use the threshold function (unit step function) so that the output will have only two possible values.

4. Plot the clustered data set using a 2D-colormap graph where the color indicates the predicted output and the x- and y-axes are the features. 


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the Iris dataset
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)

# Select Setosa and Versicolor from the dataset
selected_classes = df[df.iloc[:, 4].isin(['Iris-setosa', 'Iris-versicolor'])]

# Extract feature columns (columns 0 and 2) from Xp
X = selected_classes.iloc[:, [0, 2]].values

# Apply feature scaling to X
Xp = np.copy(X)
Xp[:, 0] = (X[:, 0] - X[:, 0].mean()) / X[:, 0].std()
Xp[:, 1] = (X[:, 1] - X[:, 1].mean()) / X[:, 1].std()

# Extract target labels (column 4) and encode them as -1 (Iris-setosa) and 1 (Iris-versicolor)
y = selected_classes.iloc[:, 4].values
y = np.where(y == 'Iris-setosa', -1, 1)

# Define the Adaline training function
def adaline(X, y, lr, epochs, tolerance):
    m, n = X.shape
    theta = np.zeros(n + 1)
    cost_list = []

    for epoch in range(epochs):
        net_input = np.dot(X, theta[1:]) + theta[0]
        errors = (y - net_input)
        theta[1:] += lr * X.T.dot(errors)
        theta[0] += lr * errors.sum()
        cost = (errors ** 2).sum() / 2.0
        cost_list.append(cost)
        
        # Check for convergence
        if cost < tolerance:
            break

    return theta, cost_list, epoch + 1  # Return the number of epochs executed

# Define a threshold function (Unit step function)
def threshold(z):
    return 1 if z >= 0 else -1

# Predict using the trained Adaline model
def predict(X, theta):
    net_input = np.dot(X, theta[1:]) + theta[0]
    return np.array([threshold(z) for z in net_input])

# Train Adaline
learning_rate = 0.01
tolerance = 0.015
theta, cost_list, epochs_executed = adaline(Xp, y, lr=learning_rate, epochs=1000, tolerance=tolerance)

# Plot the mean squared deviation vs. epoch
plt.plot(range(1, len(cost_list) + 1), cost_list, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Mean Squared Deviation')
plt.title('Mean Squared Deviation vs. Epoch')
plt.grid()
plt.show()

print(f"Algorithm converged in {epochs_executed} epochs")
```

```python
X = Xp  # Features from the dataset
y_predicted = predict(X, theta)  # Predictions from the Adaline model

# Create a colormap scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(X[y_predicted == 1][:, 0], X[y_predicted == 1][:, 1], c='b', marker='x', label='Iris-versicolor')
plt.scatter(X[y_predicted == -1][:, 0], X[y_predicted == -1][:, 1], c='r', marker='o', label='Iris-setosa')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Clustered Data (Predicted)')
plt.legend()

# Add a colormap for the data points
cbar = plt.colorbar()
cbar.set_label('Predicted Class', rotation=270, labelpad=15)
plt.show()
```
