**Train your perceptron using the data to determine the boundary line between the two classes Setosa and Versicolor.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Define the perceptron training function
def perceptron(X, y, lr, epochs):
    m, n = X.shape
    theta = np.zeros((n + 1, 1))
    n_miss_list = []

    for epoch in range(epochs):
        n_miss = 0
        for idx, x_1 in enumerate(X):
            x_i = np.insert(x_1, 0, 1).reshape(-1, 1)
            y_hat = step_func(np.dot(x_i.T, theta))

            if np.squeeze(y_hat - y[idx]) != 0:
                theta += lr * ((y[idx] - y_hat) * x_i)
                n_miss += 1

        n_miss_list.append(n_miss)

    return theta, n_miss_list

# Define the step function 
def step_func(z):
    return 1 if z > 0 else 0

# Define the plot_decision_boundary function
def plot_decision_boundary(X, theta):
    x1 = [min(X[:, 0]), max(X[:, 0])]
    m = -theta[1] / theta[2]
    c = -theta[0] / theta[2]
    x2 = m * x1 + c

    fig = plt.figure(figsize=(10, 8))
    # Plot data points for 'Setosa' (0) and 'Versicolor' (1)
    plt.plot(X[y == 0][:, 0], X[y == 0][:, 1], 'ro', label='Setosa')
    plt.plot(X[y == 1][:, 0], X[y == 1][:, 1], 'bx', label='Versicolor')
    plt.xlabel('Sepal Length')
    plt.ylabel('Petal Length')
    plt.title('Using Perceptron to Determine the Boundary Line Between Setosa and Versicolor.')
    plt.plot(x1, x2, 'y-')
    plt.legend()
    plt.show()

# Load the Iris dataset 
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)

# Select Setosa and Versicolar from the dataset
# 2 ways to do this:
# selected_classes = df[df[4].isin(['Iris-setosa', 'Iris-versicolor'])]
selected_classes = df[df[4].isin(['Iris-setosa', 'Iris-versicolor'])]

# Select columns 0 and 2 as features
X = selected_classes.iloc[:, [0, 2]].values

# Select column 4 as target labels and convert them from strings to numerical values
y = selected_classes.iloc[:, 4].map({'Iris-setosa': 0, 'Iris-versicolor': 1}).values

# Train the perceptron
theta, n_miss_list = perceptron(X, y, 0.5, 100)

# Print the learned weights and number of misclassifications
print("Learned weights:", theta)
print("Number of misclassifications per epoch:", n_miss_list)

# Plot the decision boundary
plot_decision_boundary(X, theta)
```

**Input** (X, x-1): 'X' is passed in the perceptron function as 'x-1'. There are three inputs: Bias term, sepal length and petal length.

**Weights** (theta): is initialized as a column vector of zeros in the perceptron function and is updated during training. There are three weights: Bias weight, weight for sepal length and weight for petal length.

**Net/weighted sum** (np.dot(x_i.T, theta)): is the dot product of the input vector 'X' and weight vector 'theta' then adding the bias term.

**Transfer/activation function** (step_func(z)): it takes the net sum z as input and returns 1 if z is greater than 0, otherwise, it returns 0. This function determines whether the perceptron activates (outputs 1) or not (outputs 0) based on the net sum.
